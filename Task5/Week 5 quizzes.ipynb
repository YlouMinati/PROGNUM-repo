    Write an ASCII file with your text editor with 10 rows and two columns with arbitrary numbers.

    Calculate a new array: y2 = (y - y_med)**2 where y_med is the median of y and print the contents of y2


```python
import numpy as np
from pathlib import Path
import os

data= np.loadtxt('myfile.txt')
print(data)

x, y = np.loadtxt('myfile.txt', unpack=True)

# Print min, max, and mean for x
print("Array x:")
print(f"  Min: {x.min()}")
print(f"  Max:, {x.max()}")
print(f"  Mean:, {x.mean()}")

# Print min, max, and mean for y
print("Array y:")
print(f" Min:, {y.min()}")
print(f"  Max:, {y.max()}")
print(f"  Mean:, {y.mean()}")

# Calculate median of y
y_med = np.median(y)

# Calculate new array y2 = (y - y_med)**2
y2 = (y - y_med)**2

# Print y2
print("Array y2 (squared deviation from median):")
print(y2)
```

```python
#5.10
import numpy as np
import imageio.v2 as im
from imageio.v2 import imwrite, imread
from matplotlib.pyplot import figure, show, cm

data = np.loadtxt("ngc6946.txt")

# Reschaping
cube = data.reshape(40, 50, 70)

# Midiums
y_mid = 50 // 2
x_mid = 70 // 2

z_profile = cube[:, y_mid, x_mid]

print(z_profile)

# Elementary statistics

z_std= np.std(z_profile)
z_mean= np.mean(z_profile)
z_sum= np.sum(z_profile)
z_min= np.min(z_profile)
z_max= np.max(z_profile)


#Making list
minilijstje= (
    f"statistics:\n"
)
lijstje= (
    f'statistics:\n'
    f'std = {z_std:.2f}\n'
    f'mean = {z_mean:.2f}\n'
    f'max = {z_max:.2f}\n'
    f'min = {z_min:.2f}\n'
    f'sum = {z_sum:.2f}\n'
)

print(lijstje)

plt.figure(figsize= (20, 20))
plt.plot(z_profile, marker= "$\heartsuit$", markersize= 51, color='purple', linestyle= '-')
plt.title('cube dupe', fontsize= 200, color= 'red' )
plt.gca().set_facecolor("aqua")
plt.grid(color='orange', linewidth= 20)
plt.text(0.75, 0.70, lijstje, transform=plt.gca().transAxes, verticalalignment= 'top', bbox=dict(boxstyle='round,pad=0.5', facecolor='green' , edgecolor='aqua'), fontsize= 20)
plt.show()           
```

```python
#5.11


data = np.loadtxt("ngc6946.txt")

# Reschaping
cube = data.reshape(40, 50, 70)

file_a= np.array(cube)
plane= file_a[9, :, :]

mean_p= np.mean(plane)
std_p= np.std(plane)
levels= [mean_p,mean_p + std_p, mean_p + 2*std_p]

# plotting and trotting
plt.imshow(plane)
plt.colorbar()
plt.contour(plane, levels = levels, colors= 'white')
plt.show()
```
