```python
# Script: numpy_mandelbrotComplex.py
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(h, w, maxit=20):
    '''Returns an image of the Mandelbrot fractal of size (h,w).'''
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z  = z**2 + c
        diverge = z*np.conj(z) > 2**2         # who is diverging
        div_now = diverge & (divtime==maxit)  # who is diverging now
        divtime[div_now] = i                  # note when
        z[diverge] = 2                        # avoid diverging too much

    return divtime



# Continued
fig = plt.figure()
frame = fig.add_subplot(1,1,1)
frame.imshow(mandelbrot(400,400))

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Golden ratio
phi = (1 + np.sqrt(5)) / 2

# Angle values
theta = np.linspace(0, 4*np.pi, 1000)

# Golden spiral equation
r = phi**(theta/(np.pi/2))

# Convert polar to Cartesian
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot
plt.plot(x, y)
plt.title("Golden Spiral")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
from astropy.cosmology import Planck18
import scipy as sc

# Defining constants
H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0


# Creating formula
def Tl(z):
    H_z= H0*(Om0*(1+z)**3 + Ode0)**0.5
    Tl= (1/((1+z)*H_z))

    return Tl

    
z= 2
res= sc.integrate.quad(Tl, 0, z)

print(res)

# converting to Gyr
Mpc_km = 3.08567758e19


sec_to_Gyr = 1/(3.154e7 * 1e9)

r_real= (res[0]*sec_to_Gyr)*Mpc_km

print(r_real)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate Gaussian random arrays
X = np.random.normal(0, 1, 100000)   # μ=0, σ=1
Y = np.random.normal(2, 1, 100000)   # μ=2, σ=1

# Probability density function using KDE
density_x = gaussian_kde(X)
density_y = gaussian_kde(Y)

# Range for plotting the PDFs
x_vals = np.linspace(-6, 8, 500)

pdf_x = density_x(x_vals)
pdf_y = density_y(x_vals)

# Plot histograms
plt.hist(X, bins=30, density=True, alpha=0.5,edgecolor='purple', linewidth=1, color='hotpink', label='X (μ=0, σ=1)')
plt.hist(Y, bins=30, density=True, alpha=0.5,edgecolor='purple', linewidth=1, color='aqua', label='Y (μ=2, σ=1)')

# Plot PDFs
plt.plot(x_vals, pdf_x, color='red', linestyle= '--', linewidth=2, label='PDF of X')
plt.plot(x_vals, pdf_y, color='blue', linestyle= '--', linewidth=2, label='PDF of Y')

plt.xlabel("Values")
plt.ylabel("Density")
plt.xlim(-6, 8)
plt.title("Histogram and PDF of X and Y")
plt.legend()
plt.grid(axis='y', linestyle='--', color='purple', alpha=0.7)
plt.text(-5,0.20, r"$\mathcal{N}(0,1) \rightarrow$", size= 15, color= 'red')
plt.text(4.5,0.20, r"$\leftarrow \mathcal{N}(2,1)$",size= 15, color= 'blue')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate Gaussian random arrays
X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

# ----- 2D Histogram -----
plt.figure(figsize=(7,6))

hist = plt.hist2d(X, Y, bins=30, cmap='Blues')

plt.colorbar(label="Counts")

# ----- KDE for contour -----
values = np.vstack([X, Y])
kde = gaussian_kde(values)

# Grid
x = np.linspace(-4.5, 4.5, 200)
y = np.linspace(-2.5, 6, 200)
Xgrid, Ygrid = np.meshgrid(x, y)

positions = np.vstack([Xgrid.ravel(), Ygrid.ravel()])
Z = np.reshape(kde(positions), Xgrid.shape)

# Normalize KDE to cumulative probability
Z_flat = Z.flatten()
Z_sort = np.sort(Z_flat)[::-1]
cdf = np.cumsum(Z_sort)
cdf /= cdf[-1]

# σ probability levels
levels = []
for p in [0.68, 0.95, 0.997]:
    idx = np.searchsorted(cdf, p)
    levels.append(Z_sort[idx])


contours = plt.contour(Xgrid, Ygrid, Z, levels=sorted(levels), colors='red', linewidths=2)

# Labels for contours
plt.clabel(contours, fmt={levels[0]:'1σ', levels[1]:'2σ', levels[2]:'3σ'}, colors='red')


plt.title("2D Histogram + Contour")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(-4.5, 4.5)
plt.ylim(-2.5, 6)
plt.grid(True, linestyle='--', color='black', alpha=0.6)
plt.show()
```
