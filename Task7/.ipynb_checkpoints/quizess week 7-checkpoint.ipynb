```python
#7.1

import numpy as np
from numpy import linalg as la
from numpy.linalg import *



M = np.array([1,2,3,0,-4,1,0, 3,-1]).reshape(3,3)
print("M=")
print(M)

b = np.array([2,2,1])



det = la.det(M)
print(f"Determinant of M is: {det:.2f}")


X = la.solve(M, b)  
print(M, ".(x,y,z) =", X)
x, y, z = X         
print(f"\nSolution x, y, z={x:.2f}, {y:.2f}, {z:.2f}")


print(f"\n Matrix M:\n{M}")
print(f"\nM.MT should be the identity matrix:\n{M@M.T}\n{M@M.T == np.identity(3)}")
print(f"\nInverse of A:\n{inv(M)}")
print(f"\nis equal to the transpose:\n{M.T}")
print(f"\nAlternative check with function array_equal(): {np.array_equal(M.T, inv(M))}")
print(f"\nAlternative check with function allclose(): {np.allclose(M.T, la.inv(M))}")
print(f'\nso it is orthoganol\n ')



Lambda, v = la.eig(M)
print("Lambda = ", Lambda)
print(f"\nEigenvectors:\n{v}")



# Unpack the three vectors
v1, v2, v3 = v.T
print(v1)
print(v2)
print(v3)
print("\nProof of (A.v == Lambda.v):")
print(np.allclose(M@v1, Lambda[0]*v1))
print(np.allclose(M@v2, Lambda[1]*v2))
print(np.allclose(M@v3, Lambda[2]*v3))

```

```python
#7.3
import numpy as np
from numpy import linalg as la
from numpy.linalg import *


#properties of rotation matrices are: 
#1.E^t*E^-1=I and 2. det(E)=1

from math import cos, sin

E = np.array([
    [-0.06698874, -0.87275577, -0.48353891],
    [ 0.49272847, -0.45034696,  0.74458463],
    [-0.86760081, -0.1883746,   0.46019978]
])

identity_matrix = np.eye(3)
needstobe_I = np.dot(E.T, E)
if np.allclose(needstobe_I, identity_matrix):
    print("Condition 1 (Orthogonal E^T * E == I): True")
else:
    print("Condition 1 (Orthogonal E^T * E == I): False")

det = la.det(E)
is_det_one = np.isclose(det, 1.0, atol=1e-6)

if is_det_one == 1:
    print("Condition 2 (Det == 1): True")
else:
    print("Condition 2 (Det == 1): False")
    
#inverse
minilambda, v = la.eig(E)
print("Lambda = ", minilambda)


d=np.radians(27.4)
a=np.radians(192.25)
r=np.array([cos(d)*cos(a),cos(d)*sin(a),sin(d)])
print(r)

v = np.array([-0.86760081, -0.1883746, 0.46019978])

if np.allclose(r, v):
    print("True")


lenght_v= np.sqrt((0.86760081**2)+(0.1883746**2)+(0.46019978**2))
print(f'lenght of v: {lenght_v:.2f}')

v_2 = E@v
print(v_2)
```

```python
#7.5
import numpy as np
from numpy import linalg as la
from numpy.linalg import *
import matplotlib.pyplot as plt


x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

# generates a Vandermonde matrix from the vector x (polynomial chain of al x[i])
V = np.vander(x, N=len(x), increasing=True)


# Solve for the coefficients
welp = np.linalg.solve(V, y)


print("Coefficients of the interpolating polynomial:")
print(welp)


# Create smooth x-values for plotting the polynomial
x_smooth = np.linspace(min(x), max(x), 500)

# Evaluate the polynomial at these points
y_smooth = sum(welp[i] * x_smooth**i for i in range(len(welp)))


plt.figure(figsize=(8,5))
plt.plot(x_smooth, y_smooth, label='Interpolating Polynomial', color='Hotpink')
plt.scatter(x, y, color='blue', label='Data Points', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Interpolation (Degree 9)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7.5 deel 2

# Example: you can replace these with any arrays of the same length


x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])



n = len(x)  # number of points
degree = n - 1  # degree of interpolating polynomial



# Build Vandermonde matrix 
V = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        V[i, j] = x[i] ** j  # row i, column j is x_i^j

        
        
# Solve for polynomial coefficients 
a = np.linalg.solve(V, y)



#  Create smooth x-values for plotting 
x_smooth = np.linspace(min(x), max(x), 500)



#  Evaluate the polynomial at smooth points using loops 
y_smooth = np.zeros_like(x_smooth)
for i in range(len(a)):          # loop over coefficients
    for j in range(len(x_smooth)):  # loop over smooth x-values
        y_smooth[j] += a[i] * (x_smooth[j] ** i)

# Plot 
plt.figure(figsize=(8,5))
plt.plot(x_smooth, y_smooth, label='Interpolating Polynomial', color='Hotpink')
plt.scatter(x, y, color='blue', label='Data Points', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Polynomial Interpolation (Degree {degree})')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7.6 deel 2


import numpy as np
from matplotlib.pyplot import figure, show
from scipy import interpolate as ip
from scipy.interpolate import interp1d




class Interpolation:
    def __init__(self, x, y):
        """
        Initialize with data points.
        """
        self.x = np.array(x, dtype=float)
        self.y = np.array(y, dtype=float)
        self.n = len(self.x)

        if self.n != len(self.y):
            raise ValueError("x and y must have the same length.")

    # --- Linear interpolation ---
    def linear(self):
        return interp1d(self.x, self.y, kind='linear')

    # --- Polynomial interpolation (degree n-1) ---
    def polynomial(self):
        # Build Vandermonde matrix
        V = np.vander(self.x, N=self.n, increasing=True)

        # Solve for coefficients
        coeffs = np.linalg.solve(V, self.y)

        # Evaluation function
        def poly_func(x_eval):
            scalar_input = np.isscalar(x_eval)

            x_eval = np.atleast_1d(x_eval).astype(float)

            # Vandermonde matrix for evaluation points
            V_eval = np.vander(x_eval, N=len(coeffs), increasing=True)

            y_eval = V_eval @ coeffs

            return y_eval[0] if scalar_input else y_eval

        return poly_func

    # --- Cubic spline interpolation ---
    def spline(self):
        return interp1d(self.x, self.y, kind='cubic')

    # --- General user interface ---
    def interpolate(self, x_new, method='linear'):
        """
        Interpolate value(s) at x_new using chosen method.
        
        Parameters:
            x_new : float or array-like
            method : 'linear', 'polynomial', 'spline'
        """

        x_new = np.array(x_new, dtype=float)

        # Check range
        if np.any(x_new < np.min(self.x)) or np.any(x_new > np.max(self.x)):
            raise ValueError("x_new must be within the range of x.")

        # Select method
        if method == 'linear':
            f = self.linear()
        elif method == 'polynomial':
            f = self.polynomial()
        elif method == 'spline':
            f = self.spline()
        else:
            raise ValueError("Method must be 'linear', 'polynomial', or 'spline'.")

        return f(x_new)


# =========================
# Example usage
# =========================

# Data
x = [1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9]
y = [0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0]

# Create object
interp = Interpolation(x, y)

# Test single value
x_new = 5.0
print("Linear:", interp.interpolate(x_new, method='linear'))
print("Polynomial:", interp.interpolate(x_new, method='polynomial'))
print("Spline:", interp.interpolate(x_new, method='spline'))

# =========================
# Plotting
# =========================

# Smooth x values
x_smooth = np.linspace(min(x), max(x), 500)

# Compute interpolations
y_linear = interp.interpolate(x_smooth, method='linear')
y_poly = interp.interpolate(x_smooth, method='polynomial')
y_spline = interp.interpolate(x_smooth, method='spline')

# Plot
plt.figure(figsize=(9, 5))

plt.plot(x_smooth, y_linear, label='Linear', linestyle='--')
plt.plot(x_smooth, y_poly, label='Polynomial (degree 9)')
plt.plot(x_smooth, y_spline, label='Cubic Spline')

plt.scatter(x, y, color='black', label='Data Points', zorder=5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolation Comparison')
plt.legend()
plt.grid(True)

plt.show()


```

```python
#7.7

import numpy as np

def bilinearIP(I, col, row): # Make it X, Y compatible
    '''I is a two dimensional array and iF,jF is the position at
       which we want to interpolate. 'iF' must be bigger or equal to 0 and
       smaller than the highest index of I in i. The same applies for jF'''
    rowlen, collen= I.shape
    i = int(row)
    j = int(col)
    if row - i == 0 and col - j == 0:  # Falls exactly on a gridpoint
        return I[i,j]

    inbounds = (0 <= j <= collen-1) and (0 <= i <= rowlen-1)
    if not inbounds:
        Iintp = None
    else:
        mu =  row - i
        Lambda = col - j
        # If we are at the last pixel at the right or bottom, then there
        # is not a neighbour, so we take the pixel itself as the neighbour
        I0 = I[i,j]
        I1 = I[i,min(j+1,collen-1)]
        I2 = I[min(i+1,rowlen-1),j]
        I3 = I[min(i+1,rowlen-1),min(j+1,collen-1)]
        Iintp = (1-Lambda)*(1-mu)*I0 + Lambda*(1-mu)*I1 + mu*(1-Lambda)*I2 + Lambda*mu*I3
    return(Iintp)





I = np.array([[2.0,4.1], [2.5,3.8]])
print(I)

row_y = 0.2 # Row position is y value
col_x = 0.7  # Column position is x value

Iintp = bilinearIP(I, col_x, row_y)
if Iintp:
    print(f"The interpolated intensity at i,j={row_y},{col_x} => x, y={col_x}{row_y} is {Iintp:.6f}")
else:
    print("One of the indices was outside the limits")

```
