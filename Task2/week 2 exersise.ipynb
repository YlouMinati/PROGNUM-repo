Use Python sequence type set to find values that are common to both lists (you have to convert the lists to Python sets first) with method intersection() and return those values as a list.


Use the in operator to find whether 5 is an element of the intersection list.
Use operator in to prove that 5 is a member of the set version of night1.

```python
# lists with a number of gamma ray bursts per night for two different periods:


night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

# turn into sets
n_night1 = set(night1)
n_night2 = set(night2)

# intersection
inters = set.intersection(n_night1, n_night2)

# Return to list
n_inters = list(inters)

print(n_inters)



if 5 in n_inters:
    print('yes')

else:
    print('no, there is no 5 in intersection')
    
if 5 in night1:
    print('5 is located in night1')

```

Create a NumPy array with 100000 random numbers from a Gaussian distribution with a mean (0.0) and specified standard deviation (1.4) 

and then calculates the standard deviation of the generated data.

```python
import numpy as np

mu, sigma = 0, 1.4 # mean and standard deviation

s = np.random.normal(mu, sigma, 100000000)

#print(s, "\n\n")

std = np.std(s)

print(std)

mean = np.mean(s)

print(mean)

print("\n\nDifferences\n\n")

print(np.abs(sigma-std))
print(np.abs(mu-mean))
```
